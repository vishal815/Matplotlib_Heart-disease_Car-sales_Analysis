Creating a vertical bar chart
Approach:

1 Importing matplotlib.pyplot as plt,
2 Creating list x for discrete values on x-axis,
3 Creating list y  consisting only numeric data for discrete values on y-axis,
4 Calling plt.bar() function with parameters x,y as plt.bar(x,y),
5 Setting x_label() and y_label(),
6 Setting title() for our bar chart,
7 Calling plt.show() for visualizing our chart

```python
# import matplotlib.pyplot as plt
 
x=['one', 'two', 'three', 'four', 'five']
 
# giving the values against
# each value at x axis
y=[5, 24, 35, 67, 12]
plt.bar(x, y)
 
# setting x-label as pen sold
plt.xlabel("pen sold")
 
# setting y_label as price
plt.ylabel("price")  
plt.title(" Vertical bar graph")
plt.show()
```

Creating a horizontal bar chart
Approach:

1 Importing matplotlib.pyplot as plt,
2 Creating list y   for discrete values on y-axis,
3 Creating list x  consisting only numeric data for discrete values on x-axis,
4 Calling plt.barh() function with parameters y,x as plt.barh(y,x),
5 Setting x_label() and y_label(),
6 Setting title() for our bar chart,
7 Calling plt.show() for visualizing our chart

```python
# import matplotlib.pyplot as plt
y=['one', 'two', 'three', 'four', 'five']
 
# getting values against each value of y
x=[5,24,35,67,12]
plt.barh(y, x)
 
# setting label of y-axis
plt.ylabel("pen sold")
 
# setting label of x-axis
plt.xlabel("price")
plt.title("Horizontal bar graph")
plt.show()
```

## Which one shouldyou use? (pyplot vs matplotlib OO method?)
* When plotting  somthing quickly, use pyplot method
* When plotting somthing more advanced, use OO method

```python
heart_disease.head()
```

```python
over_50=heart_disease[heart_disease["age"]>50]
len(over_50)
```

```python
over_50=heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind='scatter',
       x='age',
       y='chol',
       c='target');     #c=color
```

```python
# OO method mixed with pyplot method
fix, ax=plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
       x='age',
       y='chol',
       c='target',
       ax=ax);     

#ax.set_xlim([45,100])  
```

```python
#  OO method from scratch
fig, ax= plt.subplots(figsize=(10,6))

#plotthe data
scatter= ax.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"]);

#Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
         linestyle='--' );
```

```python
over_50.head()
```

```python
# Subplot of chol , age, thalach
fig, (ax0, ax1)=plt.subplots(nrows=2,
                            ncols=1,
                            figsize=(10,10))
# Add data to ax0
scatter =ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])

# Custmize ax0
ax0.set(title="Heart Desease and Cholesterol Levels",
#       xlabel="Age",
      ylabel="Cholesterol");

# add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

#add a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle='--');

#add data to ax1
scatter =ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"])
#costomize ax1
ax1.set(title="Heart Desease and Max Heart Rate",
      xlabel="Age",
      ylabel="Max Heart Rate")

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

#Add a meanline 
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle='--');

#add a title to figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
